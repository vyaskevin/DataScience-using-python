```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes=fig.add_axes([0.01,0.01,0.5,0.5])
axes.plot(x,y,'b')
axes.set_xlabel('Set x label')
axes.set_ylabel('Set y label')
axes.set_title('Set title')
```

```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])

axes1.plot(x,y,'b--')
axes1.set_xlabel('Set x label')
axes1.set_ylabel('Set y label')
axes1.set_title('Set title')

axes2.plot(y,x,'r*-')
axes2.set_xlabel('Set x label')
axes2.set_ylabel('Set y label')
axes2.set_title('Set title')
```

```python
np.random.seed(1)
df=pd.DataFrame(np.random.randint(1,100,(3,3)),index=['Naitik','Teerth','Fenil'],columns=['Python','Data Science','DBMS'])
df
```

```python
fig1, ax1=plt.subplots(2,3, sharex=True, sharey=True, layout='constrained')

ax1[1,2].plot(x,y)
ax1[1,2].set_title('Last')
ax1[1,2].set_ylabel('Y axis')
ax1[0,1].plot(x,x)
```

```python
fig2, ax2=plt.subplots(2,2,layout='constrained',sharex=True, sharey=True,)

ax2[0,0].plot(df.index,df.Python,'r-.',label='Python',lw='0.75',marker='s')
ax2[0,0].legend()
ax2[0,0].set_title('Python Marks')
ax2[0,0].set_xlabel('Marks')

ax2[0,1].plot(df.index,df.DBMS,'b--',label='DBMS',lw='0.25',marker='o')
ax2[0,1].legend()
ax2[0,1].set_title('DBMS Marks')

ax2[1,0].plot(df.index,df['Data Science'],'g-*',label='Data Science',marker='d')
ax2[1,0].legend()
ax2[1,0].set_title('Data Science Marks')

ax2[1,1].plot(df.index,df.Python,'r-.',label='Python',lw='1.25',marker='^')#'rgbcymk' is the colour red
ax2[1,1].plot(df.index,df.DBMS,'b--',label='DBMS',lw='1.25',marker='^')
ax2[1,1].plot(df.index,df['Data Science'],'g-*',label='Data Science',lw='1.25',marker='^')

ax2[1,1].legend()
ax2[1,1].set_title('All Marks')
```

rugplot(displays individual data points along with a single axis)(Univarient)

kdeplot(estimates the probability density function of a continuous random variable)

Jointplot

jointplot() allows you to basically match up two displot for bivariate data.With your choise of what kind parameters to compare with :

scatter
reg
resid
kde
hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.pairplot(tips,hue='day',palette='Blues')
```

```python
sns.pairplot(tips,hue='sex',palette="brg")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

Categorical Data Plots

Now lets discuss using seaborn to plot categorical data! There are a few main plot types for this :

factorplot
boxplot
violinplot
stripplot
barplot
countplot

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",hue='smoker',data=tips,palette="rainbow")
```

```python
sns.boxplot(x="total_bill",hue='sex',data=tips,palette="rainbow",orient='h')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue="sex",dodge=True)
```
